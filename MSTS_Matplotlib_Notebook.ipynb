# Funcitonal Methodology
We can use matplotlib using the basic predefined matlplotlib functions as we see below. This is very handy and you can do alot with this.

```python
#Functional way
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis Title Here')   #Add x axis label
plt.ylabel('Y Axis Title Here')   #Add y axis label
plt.title('Chart Title Here')    #Add title label
plt.show()
#Use plt.show to "print" the plot
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2) #Now you are specifying that you want the second plot number
plt.plot(y, x, 'g*-');
```

# Object Oriented Method
We can also use the OOP method to instatiate figure objects and then call methods or attributes on that object. Same results but one may be preferred to the other. If you are looking to plot something super quickly to get a quick look, the functional methods may be for you. If you are looking to make nice presentable plots, you will probably prefer to use OOP.


```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 1, 1]) # left, bottom, width, height (range 0 to 1)
#First two floats will be useful when multiple plots are done

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
#Think of a figure as a collection of plots
fig = plt.figure()

#Whenever we want to add a plot we need to add axes for that plot
#We can add two for example
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([1, 0.1, 0.4, 0.3]) # inset axes #play around with values here

#First two parameters control where in the figure the plot will be
#Second two parameters control the size of the plot

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
#We can create a set of subplots directly using the following commands
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig  
#Used to make sure charts dont overlap
plt.tight_layout()
```

A common issue with matplolib is overlapping subplots or figures. We ca use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

# Figure size , aspect ratio and DPI
Matplotlib allows the aspect ratio, DPI and figure size to be specified when Figure object is created:

* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig = plt.figure(figsize = (3,2), dpi = 100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize = (3,2), dpi = 100)
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## Setting colors, linewidths, linetypes

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes. 

The plot functin can take in a number of inputs where you can specify what you want your chart to look like: 

* color
* linewidth
* alpha
* linestyle 
* marker
* markersize
* markerfacecolor
* markeredgewidth
* markeredgecolor

Use these to customize your chart in whatever way you prefer

                                       

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(y,x, color = '#FF8C00', 
        marker = 'o', 
        markersize = 30,
        markerfacecolor = "blue",
        markeredgewidth = 10,
        markeredgecolor = 'green'
       )

ax.plot(x,y,color='blue', 
        linewidth = 3, 
        alpha = 0.5, 
        linestyle=':')

# Set your lower and upper bounds
ax.set_xlim([0,5])

#Colors can be common words, Hex codes and more
#Use linewidth to choose width (lw)
#Use alpha for transparency
#Use linestyle (ls)
```

```python
#Here is a very complete example I found online which you can refer to as insipration

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
